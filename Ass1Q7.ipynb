```python
import matplotlib.pyplot as plt
```

```python
a=input("Enter the string : ")
D={}
def Count(x):
    s=x.lower()
    for i in  x:
        if i in D:
            D[i]+=1
        else:
            D[i]=1
    
Count(a)
plt.xlabel("Characters")
plt.ylabel("Frequency")
plt.title("Character frequency in string")
plt.plot(D.keys(),D.values())
plt.show()
```
